````python
import kagglehub
import os

path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")
print("Path to dataset files:", path)

files = os.listdir(path)
print(files)
````

````python
import pandas as pd

df_orders = pd.read_csv(os.path.join(path, "olist_orders_dataset.csv"))
df_products = pd.read_csv(os.path.join(path, "olist_products_dataset.csv"))
df_products_category = pd.read_csv(os.path.join(path, "product_category_name_translation.csv")) 
df_items = pd.read_csv(os.path.join(path, "olist_order_items_dataset.csv"))
df_sellers = pd.read_csv(os.path.join(path, "olist_sellers_dataset.csv"))
df_customers = pd.read_csv(os.path.join(path, "olist_customers_dataset.csv"))
df_reviews = pd.read_csv(os.path.join(path, "olist_order_reviews_dataset.csv"))
df_payments = pd.read_csv(os.path.join(path, "olist_order_payments_dataset.csv"))
df_locations = pd.read_csv(os.path.join(path, "olist_geolocation_dataset.csv"))

display(df_orders.head(3))
display(df_products.head(3))
display(df_reviews.head(3))
display(df_locations.head(3))
````

````python
import sqlite3
import pandas as pd

# Connect to SQLite (in-memory) to create database
conn = sqlite3.connect(':memory:')

# Convert dataframe to sql table
df_orders.to_sql('orders', conn, index=False, if_exists='replace')
df_products.to_sql('products', conn, index=False, if_exists='replace')
df_products_category.to_sql('products_category', conn, index=False, if_exists='replace')
df_items.to_sql('items', conn, index=False, if_exists='replace')
df_sellers.to_sql('sellers', conn, index=False, if_exists='replace')
df_customers.to_sql('customers', conn, index=False, if_exists='replace')
df_reviews.to_sql('reviews', conn, index=False, if_exists='replace')
df_payments.to_sql('payments', conn, index=False, if_exists='replace')
df_locations.to_sql('locations', conn, index=False, if_exists='replace')

# Query using SQL. read_sql() → executes SQL query
query = "SELECT * FROM orders LIMIT 5;"
pd.read_sql(query, conn)
````

````python
query = """
SELECT order_status, COUNT(*) AS n
FROM orders
GROUP BY order_status
HAVING COUNT(*) > 100;
"""
pd.read_sql(query, conn)
````

# JOIN

````python
# INNER JOIN
query = """
SELECT o.order_id, c.customer_city
FROM orders o
JOIN customers c
ON o.customer_id = c.customer_id;
"""
pd.read_sql(query, conn)
````

````python
# LEFT JOIN
# Left table = orders. Keeps all orders. Adds customer info if available
# Every row = one order. If customer missing → customer_city = NULL
# Use case: Analyze transactions (order-centric view)
query = """
SELECT o.order_id, c.customer_city
FROM orders o
LEFT JOIN customers c
ON o.customer_id = c.customer_id;
"""
pd.read_sql(query, conn)
````

````python
# LEFT JOIN
# Left table = customers. Keeps all customers. Adds order info if available
# Every row = one customer (but may repeat if multiple orders). If no order → order_id = NULL
# Use case: Analyze customers (customer-centric view)
query = """
SELECT o.order_id, c.customer_city
FROM customers c
LEFT JOIN orders o
ON o.customer_id = c.customer_id;
"""
pd.read_sql(query, conn)
````

## SQLite does not support RIGHT JOIN (and also does not support FULL OUTER JOIN)
RIGHT JOIN can be rewritten using LEFT JOIN

````python
# RIGHT JOIN
# มี customer คนไหนที่ ไม่มี order บ้าง
query = """
SELECT c.customer_city, o.order_id
FROM customers c
LEFT JOIN orders o
ON o.customer_id = c.customer_id;
"""
pd.read_sql(query, conn)
````

````python
# OUTER JOIN
query = """
SELECT o.order_id, c.customer_id
FROM orders o
LEFT JOIN customers c
ON o.customer_id = c.customer_id

UNION

SELECT o.order_id, c.customer_id
FROM customers c
LEFT JOIN orders o
ON o.customer_id = c.customer_id;
"""
pd.read_sql(query, conn)
````

````python
# OUTER JOIN
# แบ่งกลุ่ม match vs non-match. มี record ไหนที่ match / ไม่ match ระหว่าง orders กับ customers?
query = """
SELECT 
    CASE 
        WHEN o.order_id IS NOT NULL AND c.customer_id IS NOT NULL THEN 'matched'
        WHEN o.order_id IS NOT NULL THEN 'order_only'
        ELSE 'customer_only'
    END AS status,
    COUNT(*) AS n
FROM orders o
LEFT JOIN customers c
ON o.customer_id = c.customer_id
GROUP BY status;
"""
pd.read_sql(query, conn)
````

# CASE WHEN (conditional logic)

````python
query = """
SELECT order_id,
    CASE 
        WHEN price > 100 THEN 'high'
        WHEN price > 50 THEN 'medium'
        ELSE 'low'
    END AS price_level
FROM items;
"""
pd.read_sql(query, conn)
````

# DATE / TIME

````python
# ดึงปี / เดือน
query = """
SELECT 
    strftime('%Y', order_purchase_timestamp) AS year,
    COUNT(*) AS n
FROM orders
GROUP BY year;
"""
pd.read_sql(query, conn)
````

````python
# Filter วันที่
query = """
SELECT *
FROM orders
WHERE date(order_purchase_timestamp) >= '2018-01-01';
"""
pd.read_sql(query, conn)
````

````python
# filter orders in the AM (morning hours) between 2017-2019
query = """
SELECT *
FROM orders
WHERE strftime('%Y', order_purchase_timestamp) BETWEEN '2017' AND '2019' 
  AND CAST(strftime('%H', order_purchase_timestamp) AS INTEGER) < 12;
"""
pd.read_sql(query, conn)

# WHERE strftime('%Y', order_purchase_timestamp) BETWEEN '2017' AND '2019' = 
# WHERE order_purchase_timestamp >= '2017-01-01' AND order_purchase_timestamp < '2019-01-01'
````

# Subquery

````python
# Inner query (subquery) = Returns all customer_id values for customers located in sao paulo
# Outer query = Selects all orders where the customer_id matches any ID from that list
query = """
SELECT *
FROM orders
WHERE customer_id IN (
    SELECT customer_id
    FROM customers
    WHERE customer_city = 'sao paulo'
);
"""
pd.read_sql(query, conn)
````

````python
# Above subquery equivalent to this JOIN
query = """
SELECT o.*
FROM orders o
JOIN customers c
ON o.customer_id = c.customer_id
WHERE c.customer_city = 'sao paulo';
"""
pd.read_sql(query, conn)
````

# Window Functions

````python
query = """
SELECT 
    order_id,
    price,
    AVG(price) OVER () AS global_avg
FROM items;
"""
pd.read_sql(query, conn)
````

````python
# Ranking
query = """
SELECT 
    order_id,
    price,
    RANK() OVER (ORDER BY price DESC) AS rank_price
FROM items;
"""
pd.read_sql(query, conn)
````

# Full example #1
To identify high-activity cities and classify them into high-value vs low-value segments based on average purchase price.

````python
query = """
SELECT 
    c.customer_city,
    COUNT(o.order_id) AS n_orders,
    AVG(i.price) AS avg_price,
    CASE 
        WHEN AVG(i.price) > 100 THEN 'high_value'
        ELSE 'low_value'
    END AS segment
FROM orders o
JOIN customers c ON o.customer_id = c.customer_id
JOIN items i ON o.order_id = i.order_id
GROUP BY c.customer_city
HAVING n_orders > 50
ORDER BY avg_price DESC;
"""
result1 = pd.read_sql(query, conn)
display(result1)

# result1.to_csv("city_purchase_segmentation.csv", index=False)
````

# Full Example #2
Identify products (“menu”) ordered in AM (morning) during 2017–2019, and compute: 
- total revenue
- number of orders

````python
query = """
SELECT 
    p.product_category_name AS menu,
    COUNT(DISTINCT o.order_id) AS n_orders,
    SUM(i.price) AS revenue
FROM orders o
JOIN items i 
    ON o.order_id = i.order_id
JOIN products p 
    ON i.product_id = p.product_id
JOIN products_category pc
    ON p.product_category_name = pc.product_category_name_english
WHERE o.order_purchase_timestamp >= '2017-01-01'
  AND o.order_purchase_timestamp < '2018-01-01'
  AND CAST(strftime('%H', o.order_purchase_timestamp) AS INTEGER) < 12
GROUP BY pc.product_category_name_english
ORDER BY revenue DESC;
"""
result2 = pd.read_sql(query, conn)
display(result2)

# result2.to_csv("AMorders.csv", index=False)
````
